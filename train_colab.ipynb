```python
import sys
from google.colab import drive
drive.mount('/content/drive', force_remount=False)
```

```python
from ultralytics import YOLO
```

```python
PROJECT_PATH = '/content/drive/MyDrive/Projet_DL_ML/DroneDetection'
sys.path.append(f"{PROJECT_PATH}")

DATA_PATH = PROJECT_PATH + "/Robflow/"
YAML_PATH =  PROJECT_PATH + "/drone_dataset.yaml"
MODEL_PATH = PROJECT_PATH + "/models"
MODEL_NAME = "drone_yolov5s"
model = YOLO('yolov5su.pt')
```

```python
model.train(
    data=YAML_PATH,
    epochs=50,
    imgsz=320,
    batch=32,
    project=MODEL_PATH,
    name=MODEL_NAME,
    device='cuda',
)
```

```python
results = model.val(data=YAML_PATH)
```
